```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
df = pd.read_excel("/content/C├│pia de Exercicio_Lucro.xlsx")
df.head()
```

```python
df.corr()
```

```python
import statsmodels.formula.api as smf
```

```python
Lucro_escolar = "Lucro~Investimento+Investimento_Mkt+Gastos_Administrativos"
model = smf.ols(formula=Lucro_escolar, data=df).fit()
print(model.summary())
```
